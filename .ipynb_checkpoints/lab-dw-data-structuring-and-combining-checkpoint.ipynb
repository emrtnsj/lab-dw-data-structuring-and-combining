In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
def clean_data(df):
    # Convert column names to lowercase and replace spaces with underscores
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    
    # Rename 'st' column to 'state'
    df.rename(columns={'st': 'state'}, inplace=True)
    
    # Clean 'gender' column
    df["gender"] = df["gender"].replace({"F":"F","M":"M","Femal":"F","female":"F","Male":"M","male":"M","Female":"F","Mal":"M"})
    
    # Replace abbreviated state names with full names
    df["state"] = df["state"].replace({"Calie":"California","WA":"Washington","AZ":"Arizona"})
    
    # Standardize education categories
    df["education"] = df["education"].replace({"Bachelors":"Bachelor"})
    
    # Simplify 'vehicle_class' categories
    df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car":"Luxury","Luxury SUV":"Luxury","Luxury Car":"Luxury"})
    
    
    # Convert 'customer_lifetime_value' to numeric
    df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
    
    # Define function to extract number of complaints
    def complaints(complaint):
        if isinstance(complaint, str):  #if complaint is type string
            return int(complaint.split('/')[0])
        else:
            return complaint
    
    # Apply 'complaints' function to 'number_of_open_complaints' column
    df["number_of_open_complaints"] = df["number_of_open_complaints"].apply(complaints)
    
    # Drop rows with all null values
    df.dropna(how='all', inplace=True)
    
    # Fill null values in numeric columns with mean and convert to int
    numeric_cols = df.select_dtypes(include='int').columns
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean()).astype(int)
    
    # Fill null values in categorical columns with mode
    cat_cols = df.select_dtypes(include='object').columns
    df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
    
    return df

```

```python
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'


df1 = pd.read_csv(url1)
clean_df1 = clean_data(df1)
#display(df1)
display(clean_df1)


df2 = pd.read_csv(url2)
clean_df2 = clean_data(df2)
#display(df2)
display(clean_df2)


df3 = pd.read_csv(url3)
#display(df3)
#print(df3.dtypes)
clean_df3 = clean_data(df3)
display(clean_df3)


```

```python
#now that the data is cleaned lest concatenate
concatenated_clean_df = pd.concat([clean_df1,clean_df2,clean_df3], ignore_index = True)
display(concatenated_clean_df)

```
